```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("StressLevelDataset.csv")
```

```python
data.head()
```

```python
sns.heatmap(data.isnull())
```

```python
data.isnull().sum()
```

```python
data.drop(["anxiety_level"], axis=1,inplace=True)
data.head()
```

```python
data.self_esteem = [1 if value == "M" else 0 for value in data.self_esteem]
data.head()
```

```python
data["self_esteem"].value_counts().plot(kind="bar")
```

```python
data.info()
```

```python
data.drop(columns=["self_esteem"], axis=1, inplace=True)
```

```python
y = data["mental_health_history"]
x = data.drop(["mental_health_history"], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled= scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.30, random_state=22)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
